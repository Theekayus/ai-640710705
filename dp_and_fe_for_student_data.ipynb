# ใช้ฟังก์ชัน dropna()
เพื่อลบแถวที่คอลัมที่ไม่มีข้อมูล ข้อมูลเป็น NaN ให้เหลือเฉพาะข้อมูลที่มีค่าครบถ้วน


```python
df.head()
```

```python
print(f"\nค่าเฉลี่ยอายุ (รวมค่าผิดปกติ) - {df['อายุ'].mean():.2f}")
```

```python
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
```

```python
IQR = Q3 - Q1
```

```python
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df = df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)]
```

```python
df
```

# ใช้วิธี IQR (Interquartile Range)
เพราะค่าเฉลี่ยของอายุสูงผิดปกติ

```python
df
```

```python
faculty_mapping = {
    'ชาย': 'ชาย',
    'ช': 'ชาย',
    'M': 'ชาย',
    'Male': 'ชาย',
    'หญิง': 'หญิง',
    'ญ': 'หญิง',
    'F': 'หญิง',
    'Female': 'หญิง'}
```

```python
df['เพศ'] = df['เพศ'].map(faculty_mapping)
```

```python
df
```

```python
faculty_mapping = {
    'วิทย์': 'วิทยาศาสตร์',
    'วิทยาศาสตร์': 'วิทยาศาสตร์',
    'คณะวิทยาศาสตร์': 'วิทยาศาสตร์',
    'Sci': 'วิทยาศาสตร์',
    'ศิลปศาสตร์': 'ศิลปศาสตร์',
    'Arts': 'ศิลปศาสตร์',
    'Liberal Arts': 'ศิลปศาสตร์',
    'คณะศิลปศาสตร์': 'ศิลปศาสตร์',
    'ศิลป์': 'ศิลปศาสตร์',
    'วิศวะ': 'วิศวกรรมศาสตร์',
    'วิศวกรรม': 'วิศวกรรมศาสตร์',
    'วิศวกรรมศาสตร์': 'วิศวกรรมศาสตร์',
    'วิศวฯ': 'วิศวกรรมศาสตร์',
    }
```

```python
df['คณะ'] = df['คณะ'].map(faculty_mapping)
```

# ใช้ faculty mapping
เพื่อทำให้ข้อมูลชื่อเพศและชื่อคณะเป็นมาตรฐาน

```python
df.head()
```

```python
df.loc[df['GPA'] < 0, 'GPA'] = np.nan
df.loc[df['GPA'] > 4, 'GPA'] = 4.0
mean_gpa = df['GPA'].mean()
df['GPA'].fillna(mean_gpa, inplace=True)
```

```python
df
```

# แทนค่าที่ผิดด้วยค่าเฉลี่ย
แก้ไขค่า GPA ที่ผิดปกติ โดยการแทนค่าด้วยค่าเฉลี่ยของ GPA ทั้งหมดโดยต้องไม่เกิน 4.00

```python
print(df)
print("\n📊 สรุปปัญหาที่พบในข้อมูล:")
print(f"- ค่าหายทั้งหมด: {df.isnull().sum().sum()} ค่า")
print(f"- อายุผิดปกติ: {df[df['อายุ'] > 100]['อายุ'].values if any(df['อายุ'] > 100) else 'ไม่มี'}")
print(f"- GPA ผิดปกติ: {df[(df['GPA'] < 0) | (df['GPA'] > 4)]['GPA'].values if any((df['GPA'] < 0) | (df['GPA'] > 4)) else 'ไม่มี'}")
print(f"- รูปแบบการเขียนคณะ: {df['คณะ'].nunique()} แบบ")
print(f"- รูปแบบการเขียนเพศ: {df['เพศ'].nunique()} แบบ")
```
